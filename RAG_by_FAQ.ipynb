# Import Library

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

# Document Loader

```python
def read_excel_to_dict(file_path):
  """
  Membaca file excel dari data informasi lalu disimpan dalam bentuk dictionary,
  dengan key nya yaitu kolom pertama dan value nya kolom kedua.
  """
  df = pd.read_excel(file_path)
  df = df.drop(columns=['No'])
  data_dict = dict(zip(df.iloc[:, 0], df.iloc[:, 1]))
  return data_dict

file_path = 'Data/Data Informasi.xlsx'
data = read_excel_to_dict(file_path)
data
```

# Using Transformer for Sentence Similarity 

```python
# Mengambil list pertanyaan
sentences = list(data.keys())
sentences

model = SentenceTransformer('firqaaa/indo-sentence-bert-base')
embeddings = model.encode(sentences)

input = ["Mau mengurus surat Belum Pernah Menikah"]
# input_user = input("Tuliskan pertanyaan: ")
```

# Input Output

```python
# Hitung embedding & cosine similarity untuk kalimat input
input_embedding = model.encode(input)
cosine_scores = cosine_similarity(embeddings, input_embedding)

# Temukan indeks kalimat dengan cosine similarity tertinggi
most_similar_index = cosine_scores.argmax()

# print("Kalimat paling mirip:")
# print(sentences[most_similar_index])

value = data.get(sentences[most_similar_index])
print(value)
```

# Groq

```python
from google.colab import userdata
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

groq_api_key = userdata.get('GROQ_API_KEY')

chat = ChatGroq(
    api_key = groq_api_key,
    model_name = "mixtral-8x7b-32768"
)

system = "You are a helpful assistant."
human = "{text}"

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system), ("human", human)
    ]
)

from langchain_core.output_parsers import StrOutputParser
chain = prompt | chat | StrOutputParser()

chain.invoke(
    {"text":"Why is the sky blue?"}
)
```
